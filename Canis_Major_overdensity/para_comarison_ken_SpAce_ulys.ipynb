```python
import pandas as pd

%matplotlib inline
from matplotlib import pyplot as plt

import seaborn as sns
```

```python
paras=pd.read_csv('Sp_ace/paras_full_v2.dat',sep=' ',header=None)
paras['mod']=paras[0].str.extract('(\d+)').astype(int)
paras.head(2)
```

```python
# paras=pd.read_csv('paras',sep=' ',header=None)
# paras['mod']=paras[0].str.extract('(\d+)').astype(int)
# paras.head(2)
```

```python
brights=pd.read_csv('bright.outAAT.dat',header=None,sep=r"\s*")
brights.head(2)
```

```python
merge1=pd.merge(brights,paras,left_on=[0],right_on=['mod'])
merge1.head(2)
```

## Following plots compare the the Ken's values (x_axis) with SP_Ace results.

### Bright Sample

```python
%pylab inline
pylab.rcParams['figure.figsize'] = (15, 15)
plt.subplot(321)
plt.scatter(merge1['1_x'],merge1['1_y'])
#plt.scatter(merge_u1['1_x'],merge_u1['1_y'],c='r')
plt.subplot(322)
sns.distplot(merge1['1_x']-merge1['1_y'])
plt.subplot(323)
plt.scatter(merge1['3_x'],merge1['3_y'])
plt.subplot(324)
sns.distplot(merge1['3_x']-merge1['3_y'])
plt.subplot(325)
plt.scatter(merge1['5_x'],merge1['5_y'])
plt.xlabel('Ken_values')
plt.subplot(326)
sns.distplot(merge1['5_x']-merge1['5_y'])
plt.xlabel('Ken - SP_Ace')
```
